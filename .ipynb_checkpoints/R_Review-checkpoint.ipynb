## Common Distribution functions - Workshop 3

### Binomial Distribution
二项分布:
$$P(X = x) = \binom{size}{x}p^x(1-p)^{size-x}$$
```
dbinom(x, size, prob, log = FALSE) # 某个特定取值的概率
pbinom(q, size, prob, lower.tail = TRUE, log.p = FALSE) # 累积概率
qbinom(p, size, prob, lower.tail = TRUE, log.p = FALSE) # quantile
rbinom(n, size, prob) # 产生n个b(size,prob)的二项分布随机数
```


````r
# examples
dbinom(x = c(0, 1, 2, 3, 4, 5), size = 5, prob = 0.5)
pbinom(q = 5, size = 5, prob = 0.5)
qbinom(p = 1.0, size = 5, prob = 0.5)
````

### Hypergeometric Distribution  

一共有m个白球n个黑球，不放回取k个，其中x个是白球的概率为:  

$$P(X = x) = \frac{\dbinom{m}{x}\dbinom{n}{k-x}}{\dbinom{m+n}{k}}$$
```
dhyper(x, m, n, k, log = FALSE)
phyper(q, m, n, k, lower.tail = TRUE, log.p = FALSE)
qhyper(p, m, n, k, lower.tail = TRUE, log.p = FALSE)
rhyper(nn, m, n, k)
```

#### Workshop 3 Q2
Suppose 100 fuses there are 20 defective ones. A sample of 5 fuses are randomly selected without replacement. Let X be the number of defective fuses found.
- a. Find $P(X = 0)$
- b. Find $P(X \leq 3)$
- c. Find $E(X)$
- d. Second moment of X, $E(X^2)$
- e. Variance of $X$, $Var(X)$
- f. A probability bargraph for PMF of $X$

````r
# Solutions
# a.
dhyper(x = 0, m = 20, n = 80, k = 5)
# b.
phyper(q = 3, m = 20, n = 80, k = 5)
# c.
mu = sum(1:5 * dhyper(x = 1:5, m = 20, n = 80, k = 5))
mu
# d.
ex2 = sum((1:5)^2 * dhyper(x = 1:5, m = 20, n = 80, k = 5))
ex2
# e.
ex2 - mu ^ 2
# f. 
plot(x = 0:5, dhyper(x = 0:5, m = 20, n = 80, k = 5))
````

### Customize Distribution  

- Example (SampleLabTest Q1)
Given PMF:
$$P(X=x) = \binom{200-x}{100}2^{x-200},\qquad x = 0, 1, 2, ···,100.$$

a) Find $P(X \geq 10)$  
b) Find mean, $E(X)$  
c) Find $E(X^2)$  
d) Find $E[(X + 1)^{-2}]$

````r
# solutions

# a.
x = seq(0, 100, by = 1)
prob = choose(200-x, 100) * 2^(x-200)
sum(prob[11:101])

# b.
sum(x * prob)

# c. 
sum(x^2 * prob)

# d. 
sum((x+1)^(-2) * prob)
````
